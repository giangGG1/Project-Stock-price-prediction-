```python
import matplotlib.pyplot as plt

# Data for the models and their respective RMSE values
models = [
    "LinearRegression", "Ridge Regression", "RNN cơ bản", "RNN 3 tầng mạng",
    "RNN với Dropout", "Bi-LSTM cơ bản", "Bi-LTSM kết hợp với cơ chế Attention",
    "Bi-LTSM 3 tầng mạng", "LTSM cơ bản", "LTSM 3 tầng mạng",
    "LSTM with Global Average Pooling", "Transfomer", "transfromer 2",
    "transfromer bài báo", "ransfromer cải tiến"
]

rmse_values = [
    23.11, 23.11, 42.74, 38.56, 72.13, 57.54, 99.03, 47.8, 38.5, 43.8,
    91.55, 86.39, 79.16, 105.79, 91.06
]

# Create a bar plot
plt.figure(figsize=(14, 8))
plt.barh(models, rmse_values, color='skyblue')
plt.xlabel('RMSE')
plt.title('RMSE of Different Models')
plt.gca().invert_yaxis()  # To have the lowest RMSE at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
